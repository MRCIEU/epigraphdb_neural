```python
def run_semrep(text):
    try:
        url = "http://ieu-db-interface.epi.bris.ac.uk:8067"
        r = requests.post(
            url=url, data=text.encode("utf-8"), headers={"Content-Type": "text/plain"},
        )
        r.raise_for_status()
        res = r.content.decode().split("\n")
        return res
    except:
        print("semrep failed")
        return []
```

```python
semrep_res = run_semrep(text1)
ic(semrep_res)
```

```python
semrep_res_clean = [_ for _ in semrep_res if _ != ""]
```

```python
def get_semrep_triple(semrep_item):
    cols = {
        "sub_id": 8,
        "sub_name": 9,
        "sub_type": 11,
        "sub_gene_id": 12,
        "sub_gene_name": 13,
        "pred": 22,
        "obj_id": 28,
        "obj_name": 29,
        "obj_type": 31,
        "obj_gene_id": 32,
        "obj_gene_name": 33,
    }
    fields = semrep_item.split("|")
    res = {col_key: fields[col_idx] for col_key, col_idx in cols.items()}
    return res
```

```python
output_type_col = 5
triples = [
    get_semrep_triple(_)
    for _ in semrep_res_clean
    if _.split("|")[output_type_col] == "relation"
]
ic(triples)
```
